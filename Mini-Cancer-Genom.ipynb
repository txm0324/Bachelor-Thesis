# Interactions

```python
import pandas as pd
```

```python
pathways_df = pd.read_csv("./Networks/results/Network/drug_pathway_binary_matrix.csv", index_col=0)
gene_interaction_df = pd.read_csv("./Networks/results/Network/global_gene_interaction_matrix.csv", index_col=0)
```

```python
def count_binary_percentages(df, name):
    total_values = df.size  # Gesamtzahl der Einträge im DataFrame
    count_ones = (df == 1).sum().sum()
    count_zeros = (df == 0).sum().sum()

    percent_ones = (count_ones / total_values) * 100
    percent_zeros = (count_zeros / total_values) * 100

    print(f"\n{name}:")
    print(f"Anzahl 1-en: {count_ones} → {percent_ones:.2f}%")
    print(f"Anzahl 0-en: {count_zeros} → {percent_zeros:.2f}%")
    print(f"Gesamtanzahl Einträge: {total_values}")

# Aufruf für beide DataFrames
count_binary_percentages(pathways_df, "Pathways DataFrame")
count_binary_percentages(gene_interaction_df, "Gene Interaction DataFrame")
```
